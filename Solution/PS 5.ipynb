```python
# 1. Import Libraries
import pandas as pd
import numpy as np
```

```python
# Load the iris dataset
df = pd.read_csv('iris.csv')
df.head()
```

```python
grouped_stats = df.groupby('Species').agg({
    'SepalLengthCm': ['mean', 'median', 'min', 'max', 'std'],
    'SepalWidthCm': ['mean', 'median', 'min', 'max', 'std'],
    'PetalLengthCm': ['mean', 'median', 'min', 'max', 'std'],
    'PetalWidthCm': ['mean', 'median', 'min', 'max', 'std'],
})

grouped_stats
```

```python
# Convert 'species' to numeric values
species_map = {'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2}
df['species_num'] = df['Species'].map(species_map)

# List containing numeric values for each response
species_numeric_list = df['species_num'].tolist()
print("Numeric list for species:", species_numeric_list[:10])  # print first 10 as sample
```
