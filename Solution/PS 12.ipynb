```python
# 1. Import the required libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the Iris dataset from seaborn's inbuilt dataset
iris = sns.load_dataset('iris')

# Display first few rows of the dataset
iris.head()
```

```python
# 1a. List the columns and their types
print("Features and Their Types in the Iris Dataset:")
for column in iris.columns:
    print(f"{column}: {iris[column].dtype}")
```

```python
# 2a. Plot histograms for each feature in the dataset
iris_features = iris.drop('species', axis=1)  # Dropping 'species' column for histograms

# Create subplots for histograms
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.ravel()

for i, feature in enumerate(iris_features.columns):
    sns.histplot(iris[feature], kde=True, bins=20, ax=axes[i])
    axes[i].set_title(f"Histogram of {feature}")
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
# 3a. Plot boxplots for each feature
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.ravel()

for i, feature in enumerate(iris_features.columns):
    sns.boxplot(x='species', y=feature, data=iris, ax=axes[i])
    axes[i].set_title(f"Boxplot of {feature}")
    axes[i].set_xlabel("Species")
    axes[i].set_ylabel(feature)

plt.tight_layout()
plt.show()
```

```python
# 4a. Create a demonstration of a subplot with multiple plots in a single figure
fig, axes = plt.subplots(2, 2, figsize=(14, 12))

# Plotting histograms in the subplots
sns.histplot(iris['sepal_length'], kde=True, ax=axes[0, 0]).set_title('Sepal Length Distribution')
sns.histplot(iris['sepal_width'], kde=True, ax=axes[0, 1]).set_title('Sepal Width Distribution')
sns.histplot(iris['petal_length'], kde=True, ax=axes[1, 0]).set_title('Petal Length Distribution')
sns.histplot(iris['petal_width'], kde=True, ax=axes[1, 1]).set_title('Petal Width Distribution')

plt.tight_layout()
plt.show()
```
