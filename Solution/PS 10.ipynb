```python
# 1. Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load Titanic dataset from seaborn's inbuilt dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# Check for basic information and statistics of the Titanic dataset
titanic.info()
```

```python
# 3a. Using Seaborn to visualize patterns in the dataset

# Count plot of survival status by class
sns.countplot(x='class', hue='survived', data=titanic)
plt.title('Survival Count by Class')
plt.show()

# Age distribution by survival status
sns.histplot(data=titanic, x='age', hue='survived', multiple='stack', kde=True)
plt.title('Age Distribution by Survival Status')
plt.show()

# Check relationship between class and fare
sns.boxplot(x='class', y='fare', data=titanic)
plt.title('Fare Distribution by Class')
plt.show()

# Check for fare distribution by sex
sns.violinplot(x='sex', y='fare', data=titanic)
plt.title('Fare Distribution by Gender')
plt.show()
```
